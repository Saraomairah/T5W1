## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

c = (df[['TrafficIndexLive', 'JamsCount', 'JamsDelay', 'JamsLength', 'TrafficIndexWeekAgo', 'TravelTimeHistoric', 'TravelTimeLive']]).corr()
sns.heatmap(c, annot=True)
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df['TrafficIndexLive'].plot.hist()
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.show()
```

```python
df['JamsLength'].plot.hist()
plt.title('Distribution of Jams Length')
plt.xlabel('Jams Length')
plt.show()
```

```python
df['JamsDelay'].plot.hist()
plt.title('Distribution of Jams Delay')
plt.xlabel('Jams Delay')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.scatter(data=df, x='City', y='JamsDelay')
plt.title('Distribution of Jams Delay')
plt.xlabel('Jams Delay')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(data=df, x='City', y='JamsCount')
plt.title('Distribution of Jams Count')
plt.xlabel('Jams Count')
plt.xticks(rotation=45)
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.scatter(data=df, x='JamsCount', y='JamsDelay')
plt.title('Distribution of Jams Count by Jams Delay')
plt.xlabel('Jams Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hexbin(data=df, x='JamsCount', y='JamsDelay', gridsize=15)
plt.title('Distribution of Jams Delay by Jams Count')
plt.xlabel('Jams Count')
plt.ylabel('Jams Delay')
plt.xticks(rotation=45)
plt.show()
```
