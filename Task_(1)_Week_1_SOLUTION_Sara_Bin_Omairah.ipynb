#Dataset Description

\

## The dataset contains information about road traffic accidents with the following columns:

`Time`: Time of the accident

`Day_of_week`: Day of the week when the accident occurred

`Age_band_of_driver`: Age group of the driver

`Sex_of_driver`: Gender of the driver

`Educational_level`: Educational level of the driver

`Vehicle_driver_relation`: Relationship of the driver to the vehicle

`Driving_experience`: Driving experience of the driver

`Type_of_vehicle`: Type of vehicle involved in the accident

`Owner_of_vehicle`: Owner of the vehicle

`Service_year_of_vehicle`: Service years of the vehicle

`Area_accident_occured`: Area where the accident occurred

`Lanes_or_Medians`: Lanes or medians at the accident site

`Road_allignment`: Road alignment at the accident site

`Types_of_Junction`: Type of junction at the accident site

`Road_surface_type`: Type of road surface at the accident site

`Road_surface_conditions`: Road surface conditions at the accident site

`Light_conditions`: Light conditions at the time of the accident

`Weather_conditions`: Weather conditions at the time of the accident

`Type_of_collision`: Type of collision

`Number_of_vehicles_involved`: Number of vehicles involved in the accident

`Number_of_casualties`: Number of casualties in the accident

`Vehicle_movement`: Movement of the vehicle during the accident

`Casualty_class`: Class of casualty (driver, passenger, pedestrian)

`Sex_of_casualty`: Gender of the casualty

`Age_band_of_casualty`: Age group of the casualty

`Casualty_severity`: Severity of the casualty

`Work_of_casuality`: Occupation of the casualty

`Fitness_of_casuality`: Fitness of the casualty

`Pedestrian_movement`: Movement of the pedestrian

`Cause_of_accident`: Cause of the accident

`Accident_severity`: Severity of the accident


# Tasks

##1. Data Cleaning

### Read the dataset

```python
import pandas as pd

df = pd.read_csv('Task (1) Dataset.csv')
```

### Handle Missing Values

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(10)
```

```python
df.isna().sum()
```

```python
for col in df.columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
for col in df.columns:
  print(df[col].value_counts(), '\n')
```

```python
df.info()
```

```python
df['Age_band_of_driver'].replace('Unknown', df['Age_band_of_driver'].mode()[0], inplace=True)
df['Sex_of_driver'].replace('Unknown', df['Sex_of_driver'].mode()[0], inplace=True)
df['Educational_level'].replace('Unknown', df['Educational_level'].mode()[0], inplace=True)
df['Vehicle_driver_relation'].replace('Unknown', df['Vehicle_driver_relation'].mode()[0], inplace=True)
df['Driving_experience'].replace('unknown', df['Driving_experience'].mode()[0], inplace=True)
df['Area_accident_occured'].replace('Unknown', df['Area_accident_occured'].mode()[0], inplace=True)
df['Lanes_or_Medians'].replace('Unknown', df['Lanes_or_Medians'].mode()[0], inplace=True)
df['Types_of_Junction'].replace('Unknown', df['Types_of_Junction'].mode()[0], inplace=True)
df['Weather_conditions'].replace('Unknown', df['Weather_conditions'].mode()[0], inplace=True)
df['Type_of_collision'].replace('Unknown', df['Type_of_collision'].mode()[0], inplace=True)
df['Vehicle_movement'].replace('Unknown', df['Vehicle_movement'].mode()[0], inplace=True)
df['Casualty_class'].replace('na', df['Casualty_class'].mode()[0], inplace=True)
df['Sex_of_casualty'].replace('na', df['Sex_of_casualty'].mode()[0], inplace=True)
df['Age_band_of_casualty'].replace('na', df['Age_band_of_casualty'].mode()[0], inplace=True)
df['Casualty_severity'].replace('na', df['Casualty_severity'].mode()[0], inplace=True)
df['Work_of_casuality'].replace('Unknown', df['Work_of_casuality'].mode()[0], inplace=True)
df['Pedestrian_movement'].replace('Unknown or other', df['Pedestrian_movement'].mode()[0], inplace=True)
df['Cause_of_accident'].replace('Unknown', df['Cause_of_accident'].mode()[0], inplace=True)
df['Service_year_of_vehicle'].replace('Unknown', df['Service_year_of_vehicle'].mode()[0], inplace=True)
```

```python
for col in df.columns:
  print(df[col].value_counts(), '\n')
```

### Correct any inconsistent data entries.

```python
df['Fitness_of_casuality'].replace('NormalNormal', 'Normal', inplace=True)
```

## 2. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Visualize the distribution of key variables (e.g., Age_band_of_driver, Type_of_vehicle).

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of Age Band of Driver')
sns.countplot(data=df, x='Age_band_of_driver')
plt.xticks(rotation=45)
plt.xlabel('Age Band of Driver')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of Type of Vehicle')
sns.countplot(data=df, x='Type_of_vehicle')
plt.xticks(rotation=45)
plt.xlabel('Type of Vehicle')
plt.ylabel('Count')
plt.show()
```

### Explore relationships between variables (e.g., Age_band_of_driver vs. Accident_severity).


```python
plt.figure(figsize=(10, 6))
plt.title('Relationship between Age Band of Driver and Accident Severity')
sns.countplot(data=df, x='Age_band_of_driver', hue='Accident_severity')
plt.xticks(rotation=45)
plt.xlabel('Age Band of Driver')
plt.ylabel('Count')
plt.show()
```
